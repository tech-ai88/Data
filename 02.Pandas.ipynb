## 데이터 가공 
- 분석 목적에 맞게 데이터를 재구성하거나 형태를 바꾸는 과정
- 필터링, 컬럼 선택/재구성, 정렬 등 

### 데이터 필터링
- 필터링은 데이터에서 필요한 행/열만 선택하는 과정
    - iloc[row, column] : 위치(숫자) 기반  인덱싱 
    - loc[row, column] : 라벨(이름) 기반 인덱싱 
    - boolean indexing : 조건을 이용한 데이터 선택

#### iloc
- iloc[row, column] 
- 위치(숫자) 기반 인덱싱

```python
dye=pd.read_csv('./Data_Set/염색공정.csv', encoding='cp949')   
```

```python
# 원하는 행열 
dye.iloc[5:15, 2:4]     # 행 5 ~ 14, 열 2 ~ 3

# 모든 행 
dye.iloc[:, :4]         # 열 0 ~ 3  
dye.iloc[:, 4:]         # 열 4 ~ 
dye.iloc[:, [1, 2]]     # 열 1, 2
dye.iloc[:, :-1]        # 마지막 열 제외
dye.iloc[:, -1:]        # 마지막 열만 

# 모든 열 
dye.iloc[:3, :]         # 행 0 ~ 2  / 참고 dye.iloc[:3, ] 열 생략 가능, 행 불가
```

#### 조건기반 인덱싱 
- 조건기반 인덱싱 (boolean indexing)
    - 문법 : 데이터프레임의 조건식을 직접 사용하여 필터링
        - 예 : df[df['column'] > value]
    - 장점 : Python 코드의 기본 문법을 사용하므로 직관적이고 유연함
    - 단점 : 조건이 복잡해지면 가독성이 떨어질 수 있음

```python
# 불량 >> 품질 기준을 벗어난 데이터만 확인 
fail=dye[dye['염색색차 DE'] > 3]
fail.shape
```

```python
len(fail) / len(dye) * 100         # 불량률 계산 
```

- 불량
    - 설비별 불량률 비교
    - 공정 조건 영향 

```python
fail['설비번호'].value_counts()        # 설비별 불량 건수 (0 안나옴) / FCM06, FCM15 
```

```python
bad=dye[dye['설비번호'] == 'FCM07']     # 불량률이 가장 높음 (계산은 그룹함수)
bad.describe().T.round(2)
```

```python
good=dye[dye['설비번호'] == 'FCM06']
good.describe().T.round(2)
```

```python
# Ex1) FCM07 설비이면서 동시에 정상인 데이터만 뽑기
num=dye['설비번호'] == 'FCM07'
de=dye['염색색차 DE'] < 2
result=dye[num & de]      # [cond1 | ~cond2]
result.head()
```

```python
#  Ex2) 평균 온도보다 높게 작업된 데이터 뽑기
avg_temp=dye['진행온도'].mean()
cond=dye['진행온도'] > avg_temp
result=dye[cond]
result.shape

result.loc[:, ['설비번호', '진행온도', '지시온도']].head()
```

```python
dye=pd.read_csv("./Data_Set/염색공정.csv", encoding='cp949') 
```

```python
copy_dye.head()
```

```python
copy_dye['상태'].value_counts()
```

```python
# EX1) 생성(추가) / 파생변수 - column 추가 후 원하는 위치
minus=copy_dye['지시온도'] - copy_dye['진행온도']

copy_dye.insert(loc=copy_dye.columns.get_loc('진행온도') + 1,     # 추가 위치
                column='온도편차',    # 추가할 컬럼명
                value=minus)          # 추가할 값
copy_dye.head(2)
```

#### 삭제

```python
copy_dye.drop(columns=['온도편차', '상태'], inplace=True)
copy_dye.head(2)
```
